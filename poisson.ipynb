## Working from data that includes opening weekend info as well as opening weekend ratio to domestic total as proxy for popularity

```python
addnl=pd.read_csv('final-data.csv')
```

```python
addnl.head()
```

```python
addnl.drop(columns=['Director','Distributor','DayOfTheYear^2','GenreMap_Comedy'], inplace=True)
```

```python
addnl.set_index('the_index', inplace=True)
```

```python
addnl.dtypes
```

```python
addnl.corr()
```

```python
addnl.describe()
```

```python
addnl[addnl['ProductionBudget']<200]
```

```python
for idx, row in addnl.iterrows():
    if addnl.loc[idx,'ProductionBudget']<100:
        addnl.loc[idx,'ProductionBudget']=addnl.loc[idx,'ProductionBudget']*1000000
```

```python
sns.pairplot(addnl.drop(columns=['Wins','Genre','IntMonthOfTheYear','ReleaseDate','MpaaRating']))
```

```python
features, target = ['DomesticTotalGross','RunTime','ProductionBudget','InRelease','IntDayOfTheYear','OpeningGross','OpeningRank','OpeningRatio'], ['Nominations']
X, y = addnl[features], addnl[target]
X_train, X_test, y_train, y_test = train_test_split(addnl[features], addnl[target], test_size=0.2, random_state=42)
```

```python
model = sm.OLS(y_train,X_train)
results = model.fit()
results.summary()
```

```python
ols_preds=results.predict(X_test) 
plt.scatter(ols_preds,y_test,alpha=.1)
plt.xlabel('Predicted Value')
plt.ylabel('Actual Value')
plt.title("Linear Regression")
plt.plot(np.linspace(0,15,15),np.linspace(0,15,15))
```

```python
pm_model=sm.GLM(y_train, X_train, family=sm.families.Poisson())
pm_results=pm_model.fit()
print(pm_results.summary())
```

```python
glm_preds=pm_results.predict(X_test)
plt.scatter(glm_preds,y_test,alpha=.1)
plt.xlabel('Predicted Value')
plt.ylabel('Actual Value')
plt.title("StatsModels' GLM Poisson Regression")
plt.plot(np.linspace(0,15,15),np.linspace(0,15,15))
```

```python
pm=sm.Poisson(y_train,X_train).fit()
pm.summary()
```

```python
preds=pm.predict(X_test)
```

```python
#sns.jointplot(preds,y_test, kind='regplot')
```

```python
plt.scatter(preds,y_test,alpha=.1)
plt.xlabel('Predicted Value')
plt.ylabel('Actual Value')
plt.title("StatsModels' Poisson Regression")
plt.plot(np.linspace(0,15,15),np.linspace(0,15,15))
```
